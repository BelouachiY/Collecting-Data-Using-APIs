## Demo: How to work with database


Download database file.


```python
import requests

# URL of the file to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite"

# Path to save the downloaded file
file_path = "m4_survey_data.sqlite"

# Download the file
response = requests.get(url)

# Save the file
with open(file_path, 'wb') as f:
    f.write(response.content)

print("File downloaded successfully.")
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
sql_statement = '''
           SELECT ConvertedComp FROM master
 '''
data = pd.read_sql_query(sql_statement,conn)
data.plot(kind='hist', bins=50)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
sql_statement2 = '''
           SELECT Age FROM master
 '''
data2 = pd.read_sql_query(sql_statement2,conn)
data2['Age'].plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sql_statement3= '''
           SELECT Age, WorkWeekHrs
           FROM master
'''
data3 = pd.read_sql_query(sql_statement3 , conn)
data3.plot(kind='scatter' , x='WorkWeekHrs' , y='Age')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import matplotlib.pyplot as plt
# your code goes here
sql_statement4= '''
                     SELECT Age, WorkWeekHrs, CodeRevHrs
                     FROM master
'''
data4 = pd.read_sql_query(sql_statement4,conn)
data4.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s=data4['Age']/2)


# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
statement6 = '''
              SELECT DatabaseDesireNextYear , count(DatabaseDesireNextYear) as 'count_db' from DatabaseDesireNextYear
              GROUP BY DatabaseDesireNextYear 
              ORDER by count(DatabaseDesireNextYear) DESC 
             '''
data9 = pd.read_sql_query(statement6 , conn)
data_orde = data9.head()
plt.figure(figsize=(8, 8))
plt.pie(data_orde['count_db'], labels=data_orde['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
# LanguageDesireNextYear
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.

```python
# your code goes here
A = '''
         SELECT WorkWeekHrs, CodeRevHrs
           FROM master
           WHERE Age BETWEEN 30 AND 35
'''
data11 = pd.read_sql_query(A,conn)
median_workweekhrs = data11['WorkWeekHrs'].median()
median_coderevhrs = data11['CodeRevHrs'].median()

# Create a DataFrame for the medians
median_data = pd.DataFrame({'Median': [median_workweekhrs, median_coderevhrs]}, index=['WorkWeekHrs', 'CodeRevHrs'])

# Plot the stacked chart
median_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Respondents Aged 30-35')
plt.xlabel('Metrics')
plt.ylabel('Median')
plt.xticks(rotation=0)
plt.legend(loc='upper right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
stat = '''
        SELECT Age, ConvertedComp
        FROM master
        WHERE Age BETWEEN 45 AND 60
'''

data12 = pd.read_sql_query(stat, conn)
median_converted_comp = data12.groupby('Age')['ConvertedComp'].median()

# Plot the median 'ConvertedComp' against age
median_converted_comp.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
stat0 = '''
         SELECT MainBranch as count from master
 '''
dt = pd.read_sql_query(stat0, conn)

# Count the occurrences of each unique value in the 'count' column (aliased from MainBranch)
branch_counts = dt['count'].value_counts()

# Create a horizontal bar chart
branch_counts.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.show()
```
